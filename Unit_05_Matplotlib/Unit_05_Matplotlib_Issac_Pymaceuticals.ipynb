```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem
import matplotlib.ticker as ticker

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = pd.read_csv("Instructions/pymaceuticals/data/mouse_drug_data.csv")
clinical_trial_data_to_load = pd.read_csv("Instructions/pymaceuticals/data/clinicaltrial_data.csv")

# Read the Mouse and Drug Data and the Clinical Trial Data


# Combine the data into a single dataset
merge_mouse_clinic_data = pd.merge(clinical_trial_data_to_load, mouse_drug_data_to_load, on="Mouse ID", how="left")

# Display the data table for preview
merge_mouse_clinic_data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
tumor_response = merge_mouse_clinic_data[["Drug","Timepoint","Tumor Volume (mm3)"]].groupby(["Drug","Timepoint"]).mean()

# Convert to DataFrame
tumor_response_df = pd.DataFrame(tumor_response).reset_index()

# Preview DataFrame
tumor_response_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_sem = merge_mouse_clinic_data[["Drug","Timepoint","Tumor Volume (mm3)"]].groupby(["Drug","Timepoint"]).sem()

# Convert to DataFrame
tumor_sem_df = pd.DataFrame(tumor_sem).reset_index()

# Preview DataFrame
tumor_sem_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumor_sem_df_pivot = tumor_sem_df.pivot_table(index='Timepoint',columns='Drug',values='Tumor Volume (mm3)')
tumor_response_df_pivot = tumor_response_df.pivot_table(index='Timepoint',columns='Drug',values='Tumor Volume (mm3)')

# Preview that Reformatting worked
 #tumor_sem_df_pivot.head()
tumor_response_df_pivot.head()
```

```python
# Generate the Plot (with Error Bars)
fig, ax = plt.subplots()
ax.errorbar(tumor_response_df_pivot.index, tumor_response_df_pivot.Capomulin, yerr=tumor_sem_df_pivot.Capomulin,
          fmt="o", color="b", ls="--",
          alpha=0.5, label="Capomulin")
ax.errorbar(tumor_response_df_pivot.index, tumor_response_df_pivot.Infubinol, yerr=tumor_sem_df_pivot.Infubinol,
          fmt="o", color="r", ls="--",
          alpha=0.5, label="Infubinol")
ax.errorbar(tumor_response_df_pivot.index, tumor_response_df_pivot.Ketapril, yerr=tumor_sem_df_pivot.Ketapril,
          fmt="o", color="m", ls="--",
          alpha=0.5, label="Ketapril")
ax.errorbar(tumor_response_df_pivot.index, tumor_response_df_pivot.Placebo, yerr=tumor_sem_df_pivot.Placebo,
          fmt="o", color="g", ls="--",
          alpha=0.5, label="Placebo")
ax.set_xlim(-0.5,47)
ax.set_ylim(30, 75)
ax.set_xlabel("Timepoint")
ax.set_ylabel("Tumor Volume (mm3)")
plt.title("Tumor Response to Treatment")
plt.legend(loc="best", fontsize="small", fancybox=True)

# Save the Figure
plt.savefig("../Unit_05_Matplotlib/graphs/tumor_response_to_treatment.png")
```

```python
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
metastatic_response = merge_mouse_clinic_data.groupby(["Drug","Timepoint"])["Metastatic Sites"].mean()
# Convert to DataFrame
metastatic_response_df = pd.DataFrame(metastatic_response).reset_index()

# Preview DataFrame
metastatic_response_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
metastatic_sem = merge_mouse_clinic_data[["Drug","Timepoint","Metastatic Sites"]].groupby(["Drug","Timepoint"]).sem()
# Convert to DataFrame
metastatic_sem_df = pd.DataFrame(metastatic_sem).reset_index()

# Preview DataFrame
metastatic_sem_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
metastatic_response_pivot = metastatic_response_df.pivot_table(index='Timepoint',columns='Drug',values='Metastatic Sites')
metastatic_sem_pivot = metastatic_sem_df.pivot_table(index='Timepoint',columns='Drug',values='Metastatic Sites')

# Preview that Reformatting worked
metastatic_response_pivot.head()
#metastatic_sem_pivot.head()
```

```python
# Generate the Plot (with Error Bars)
fig, ax = plt.subplots()
ax.errorbar(metastatic_response_pivot.index, metastatic_response_pivot.Capomulin, yerr=metastatic_sem_pivot.Capomulin,
           fmt="o", color="b", ls="--",
           alpha=0.5, label="Capomulin")
ax.errorbar(metastatic_response_pivot.index, metastatic_response_pivot.Infubinol, yerr=metastatic_sem_pivot.Infubinol,
           fmt="o", color="r", ls="--",
           alpha=0.5, label="Infubinol")
ax.errorbar(metastatic_response_pivot.index, metastatic_response_pivot.Ketapril, yerr=metastatic_sem_pivot.Ketapril,
           fmt="o", color="m", ls="--",
           alpha=0.5, label="Ketapril")
ax.errorbar(metastatic_response_pivot.index, metastatic_response_pivot.Placebo, yerr=metastatic_sem_pivot.Placebo,
           fmt="o", color="g", ls="--",
           alpha=0.5, label="Placebo")

ax.set_xlabel("Timepoint")
ax.set_ylabel("Metastatic Sites")
plt.title("Metastatic Response to Treatment")
plt.legend(loc="best", fontsize="small", fancybox=True)

# Save the Figure
plt.savefig("../Unit_05_Matplotlib/graphs/metastatic_response_to_treatment.png")

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
count_of_mice = merge_mouse_clinic_data.groupby(["Drug","Timepoint"])["Mouse ID"].count()
# Convert to DataFrame
count_mice_df = pd.DataFrame(count_of_mice)
count_mice_df.reset_index(inplace=True)

# Preview DataFrame
count_mice_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mouse_count_pivot = count_mice_df.pivot_table(index='Timepoint',columns='Drug',values='Mouse ID')

# Preview the Data Frame
mouse_count_pivot.head()
```

```python
# Generate the Plot (Accounting for percentages)
fig, ax = plt.subplots()

ax.errorbar(mouse_count_pivot.index, mouse_count_pivot.Capomulin/25*100,
            fmt="o", color="b", ls="--",
            alpha=0.5, label="Capomulin")

ax.errorbar(mouse_count_pivot.index, mouse_count_pivot.Infubinol/25*100,
            fmt="o", color="r", ls="--",
            alpha=0.5, label="Infubinol")

ax.errorbar(mouse_count_pivot.index, mouse_count_pivot.Ketapril/25*100,
            fmt="o", color="m", ls="--",
            alpha=0.5, label="Ketapril")

ax.errorbar(mouse_count_pivot.index, mouse_count_pivot.Placebo/25*100,
            fmt="o", color="g", ls="--",
            alpha=0.5, label="Placebo")

x_axis = np.arange(0, 1, 0.1)
e_x = [np.exp(x) for x in x_axis]

ax.set_xlabel("Timepoint")
ax.set_ylabel("Mouse Count")
plt.title("Survival Rates")
plt.legend(loc="best", fontsize="small", fancybox=True)

# Save the Figure
plt.savefig("../Unit_05_Matplotlib/graphs/Survival_Rates.png")

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
tumor_data = tumor_response_df_pivot.pct_change(9).iloc[9]*100

# Display the data to confirm
tumor_arr = tumor_data.iloc[[0,1,3,5]]
#tumor_arr.index
tumor_data
```

```python
# Store all Relevant Percent Changes into a Tuple
fig, ax = plt.subplots()
rects = ax.bar(tumor_arr.index, tumor_arr,
               color=['green' if per < 0 else 'red' for per in tumor_arr])

# Splice the data between passing and failing drugs
plt.axhline(y=0, color = "k",linewidth=4)

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., .9*height,
                '%d' % int(height)+"%",
                ha='center', va='bottom')

# Orient widths. Add labels, tick marks, etc. 
ax.set_ylabel('% Tumor Volume Change')
ax.set_title('Tumor Change Over 45 Day Treatment')

# Use functions to label the percentages of changes
autolabel(rects)


# Call functions to implement the function calls


# Save the Figure
plt.savefig("../Unit_05_Matplotlib/graphs/Summary_Bar_Graph.png")

# Show the Figure
plt.show()
```
